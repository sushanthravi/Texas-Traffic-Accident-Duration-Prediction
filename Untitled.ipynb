```python
import pandas as pd  
import numpy as np 
df = pd.read_csv("C:/Users/Lenovo/Downloads/Data Science and Programming/Final Project/final_accidents.csv")
df.head()
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
# Filtering the DataFrame for streets containing a dash
highways = df[df['Street'].str.contains('-')]

# List of specific roads with dashes to exclude
roads_with_dash = [' Almeda-Genoa Rd', ' McNeil-Merrilltown Rd', 'Beasley-Needville Rd', 'Old Addicks-Howell Rd', 'S Pecos-La Trinidad']

# Exclude the specific roads from the highways DataFrame
highways = highways[~highways['Street'].isin(roads_with_dash)]

# Get the unique list of highways
list_highways = highways['Street'].unique()

# Create a new column 'Highways' indicating whether a street is in the list of highways
df['Highways'] = df['Street'].isin(list_highways).astype(int)
df['Sunrise_Sunset'] = df['Sunrise_Sunset'].fillna(0).map({'Day': 0, 'Night': 1})
df['Civil_Twilight'] = df['Civil_Twilight'].fillna(0).map({'Day': 0, 'Night': 1})
df['Nautical_Twilight'] = df['Nautical_Twilight'].fillna(0).map({'Day': 0, 'Night': 1})
df['Astronomical_Twilight'] = df['Astronomical_Twilight'].fillna(0).map({'Day': 0, 'Night': 1})
```

```python
df = df.drop(columns=['Unnamed: 0', 'Severity', 'Street'])
random = df.sample(n=500)
random.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import plot_tree
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Assume 'random' is your DataFrame
# Identify the categorical and numerical columns
categorical_cols = [ 'City', 'County', 'Sunrise_Sunset', 'Civil_Twilight', 
                    'Nautical_Twilight', 'Astronomical_Twilight']
numerical_cols = ['Start_Lat', 'Start_Lng', 'Distance(mi)', 'Year', 'Month', 'Day', 'Hour', 'Duration']

# Reset index to ensure alignment
random = random.reset_index(drop=True)

# Check for missing values and handle them if needed
random = random.dropna(subset=['Duration'])

# One-hot encode the categorical variables
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_categorical_data = encoder.fit_transform(random[categorical_cols])
encoded_categorical_df = pd.DataFrame(encoded_categorical_data, columns=encoder.get_feature_names_out(categorical_cols))

# Combine the encoded categorical data with the numerical data
X = pd.concat([random[numerical_cols].drop(columns='Duration'), encoded_categorical_df], axis=1)
y = random['Duration'].reset_index(drop=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree Regressor
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = tree_model.predict(X_test)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(tree_model, filled=True, feature_names=X.columns, rounded=True)
plt.title('Decision Tree')
plt.show()
```

```python
# Extract feature importances
importances = tree_model.feature_importances_

# Create a DataFrame for feature importances
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
})

# Sort the DataFrame by importance
sorted_importances = feature_importance_df.sort_values(by='Importance', ascending=False)

# Display the top 2 predictors
top_20_predictors = sorted_importances.head(20)
print(top_20_predictors)
```
